```python
import numpy as np
from numpy.linalg import eigh, norm
```

```python
A = np.array([[-5,2,3],
              [2,5,1],
              [-3,1,-5]])
```

```python
A.T
```

```python
A.T@A
```

```python
eigh(A.T@A)
```

```python
ev, V = eigh(A.T@A)
```

```python
V.T
```

```python
V[0]
```

```python
u0 = A@V[:,0]/norm(A@V[:,0])
u1 = A@V[:,1]/norm(A@V[:,1])
u2 = A@V[:,2]/norm(A@V[:,2])
```

```python
U=np.array([u0,u1,u2]).T
U
```

```python
D = np.round(U.T@A@V, decimals=5)
D
```

```python
#A = U@D@V.T is the SVD
U@D@V.T
```
